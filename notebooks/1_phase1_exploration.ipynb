# Phase 1: Data Loading & Exploration

**Goal**: Build robust dataset loader and explore labels

## What we'll do:
1. Load datasets from Phase 0 or sklearn
2. Inspect label distributions
3. Visualize class counts
4. Store dataset metadata
5. Generate comprehensive reports

```python
# Import necessary libraries
import sys
import os
sys.path.append('../src')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import json

from sklearn.datasets import load_digits, load_iris, load_wine

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['figure.dpi'] = 100

print("✓ Libraries imported")
```

## 1. Initialize Data Loader

```python
from phase1_exploration.data_loader import DataLoader

# Initialize loader
loader = DataLoader(data_dir='../data')

# Check what's available
print("Available data directories:")
for root, dirs, files in os.walk('../data'):
    level = root.replace('../data', '').count(os.sep)
    indent = ' ' * 2 * level
    print(f'{indent}{os.path.basename(root)}/' if level > 0 else f'{indent}{root}/')
    subindent = ' ' * 2 * (level + 1)
    for file in files[:5]:  # Show first 5 files
        print(f'{subindent}{file}')
    if len(files) > 5:
        print(f'{subindent}... and {len(files) - 5} more')
```

## 2. Load Datasets

```python
# Try to load Phase 0 datasets
phase0_dir = '../data/synthetic'

if os.path.exists(phase0_dir):
    print("Loading Phase 0 datasets...")
    csv_datasets = loader.load_from_directory(phase0_dir, pattern="*dataset*.csv")
    print(f"✓ Loaded {len(csv_datasets)} datasets from Phase 0")
else:
    print("Phase 0 data not found. Loading sklearn datasets...")
    # Load sklearn datasets
    loader.load_from_sklearn('digits')
    loader.load_from_sklearn('iris')
    loader.load_from_sklearn('wine')

# Display what we loaded
print("\nLoaded datasets:")
summary = loader.get_dataset_summary()
print(summary.to_string(index=False))
```

## 3. Exploratory Analysis

```python
from phase1_exploration.exploratory_analysis import ExploratoryAnalyzer

# Initialize analyzer
analyzer = ExploratoryAnalyzer()

# Analyze each dataset
reports = {}

for dataset_name in loader.list_datasets():
    print(f"\n{'='*60}")
    print(f"Analyzing: {dataset_name}")
    print(f"{'='*60}")
    
    dataset_info = loader.get_dataset(dataset_name)
    
    # Generate report
    report = analyzer.generate_dataset_report(dataset_info)
    reports[dataset_name] = report
    
    # Create visualization
    fig = analyzer.plot_comprehensive_report(dataset_info)
    plt.show()
```

## 4. Advanced Visualization

```python
from phase1_exploration.visualization import DataVisualizer

# Initialize visualizer
visualizer = DataVisualizer()

# Get all datasets
all_datasets = {name: loader.get_dataset(name) for name in loader.list_datasets()}

# Create dataset comparison
print("Creating dataset comparison...")
fig1 = visualizer.plot_class_comparison(all_datasets)
plt.show()

# Create label quality indicators for the largest dataset
if loader.list_datasets():
    largest_dataset = max(loader.list_datasets(), 
                         key=lambda x: loader.get_dataset(x)['n_samples'])
    dataset_info = loader.get_dataset(largest_dataset)
    
    print(f"\nCreating label quality indicators for {largest_dataset}...")
    fig2 = visualizer.plot_label_quality_indicators(
        dataset_info['X'],
        dataset_info['y'],
        dataset_name=largest_dataset
    )
    plt.show()
```

## 5. Save Reports and Metadata

```python
# Create reports directory
import os
os.makedirs('../reports/phase1', exist_ok=True)
os.makedirs('../reports/visualizations', exist_ok=True)
os.makedirs('../reports/metadata', exist_ok=True)

# Save visualizations
print("Saving visualizations...")
visualizer.save_visualization(fig1, 'phase1_dataset_comparison', 
                             directory='../reports/visualizations')
if 'fig2' in locals():
    visualizer.save_visualization(fig2, f'phase1_{largest_dataset}_quality_indicators',
                                 directory='../reports/visualizations')

# Save comprehensive reports
print("\nSaving comprehensive reports...")
for dataset_name, report in reports.items():
    dataset_info = loader.get_dataset(dataset_name)
    fig = analyzer.plot_comprehensive_report(
        dataset_info,
        save_path=f'../reports/phase1/phase1_{dataset_name}_report.png'
    )
    plt.close(fig)
    print(f"  ✓ {dataset_name} report saved")

# Save metadata
total_samples = sum([loader.get_dataset(name)['n_samples'] 
                    for name in loader.list_datasets()])
total_features = sum([loader.get_dataset(name)['n_features'] 
                     for name in loader.list_datasets()])

metadata = {
    'phase': 1,
    'datasets_analyzed': loader.list_datasets(),
    'total_samples': total_samples,
    'total_features': total_features,
    'timestamp': pd.Timestamp.now().isoformat(),
    'reports_generated': list(reports.keys())
}

with open('../reports/metadata/phase1_metadata.json', 'w') as f:
    json.dump(metadata, f, indent=2)

print("\n✓ Metadata saved to ../reports/metadata/phase1_metadata.json")
```

## 6. Summary Statistics

```python
print("📊 Phase 1 Summary:")
print("=" * 40)

for dataset_name in loader.list_datasets():
    dataset_info = loader.get_dataset(dataset_name)
    report = reports[dataset_name]
    
    print(f"\n{dataset_name.upper()}:")
    print(f"  • Samples: {dataset_info['n_samples']:,}")
    print(f"  • Features: {dataset_info['n_features']}")
    print(f"  • Classes: {report['label_stats']['n_classes']}")
    print(f"  • Imbalance ratio: {report['label_stats']['imbalance_ratio']:.2f}")
    print(f"  • Balanced: {'Yes' if report['label_stats']['is_balanced'] else 'No'}")

print(f"\n{'='*40}")
print(f"Total datasets: {len(loader.list_datasets())}")
print(f"Total samples: {total_samples:,}")
print(f"Total features: {total_features}")
```

## 7. Export Clean Data for Next Phase

```python
# Create directory for Phase 2
os.makedirs('../data/phase2_ready', exist_ok=True)

# Export datasets for Phase 2
print("Exporting datasets for Phase 2...")

for dataset_name in loader.list_datasets():
    dataset_info = loader.get_dataset(dataset_name)
    
    # Create DataFrame
    X_df = pd.DataFrame(dataset_info['X'], columns=dataset_info['feature_names'])
    df = X_df.copy()
    df['target'] = dataset_info['y']
    
    # Save
    filename = f'../data/phase2_ready/{dataset_name}_cleaned.csv'
    df.to_csv(filename, index=False)
    print(f"  ✓ {dataset_name}: {len(df)} samples → {filename}")

print("\n✅ Phase 1 Complete!")
print("\nNext: Phase 2 - Noise Detection (Baseline)")
```
