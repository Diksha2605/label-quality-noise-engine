# Phase 0: Environment Setup and Dataset Preparation

**Goal**: Set up environment, understand labels and noise.

## What we'll do:
1. Load clean datasets (sklearn digits, iris, wine)
2. Create synthetic datasets
3. Inject synthetic noise (random, class-dependent, hard examples)
4. Save datasets for future phases

```python
# Import necessary libraries
import sys
import os
sys.path.append('../src')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_digits, load_iris, load_wine
from sklearn.datasets import make_classification

# Set style for plots
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

print("✓ Libraries imported")
```

## 1. Load Clean Datasets

```python
from phase0_preparation.dataset_preparation import DataPreparation

# Initialize data preparation
prep = DataPreparation()

# Load sklearn datasets
sklearn_data = prep.load_sklearn_datasets()

print("Sklearn datasets loaded:")
for name, data in sklearn_data.items():
    print(f"  • {name.upper()}: {data['X'].shape[0]} samples, {data['X'].shape[1]} features, {len(np.unique(data['y']))} classes")
```

## 2. Create Synthetic Dataset

```python
# Create a synthetic dataset
synthetic_data = prep.create_synthetic_dataset(
    n_samples=1000,
    n_features=20,
    n_classes=3,
    random_state=42
)

print(f"Synthetic dataset created:")
print(f"  • Samples: {synthetic_data['X'].shape[0]}")
print(f"  • Features: {synthetic_data['X'].shape[1]}")
print(f"  • Classes: {len(np.unique(synthetic_data['y']))}")
```

## 3. Visualize Dataset Characteristics

```python
# Create a summary visualization
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Plot 1: Digits class distribution
digits_data = sklearn_data['digits']
unique_digits, counts_digits = np.unique(digits_data['y'], return_counts=True)
axes[0, 0].bar(unique_digits, counts_digits)
axes[0, 0].set_title('Digits Dataset - Class Distribution')
axes[0, 0].set_xlabel('Digit Class')
axes[0, 0].set_ylabel('Count')

# Plot 2: Iris class distribution
iris_data = sklearn_data['iris']
unique_iris, counts_iris = np.unique(iris_data['y'], return_counts=True)
axes[0, 1].bar(range(len(unique_iris)), counts_iris, tick_label=iris_data['target_names'])
axes[0, 1].set_title('Iris Dataset - Class Distribution')
axes[0, 1].set_ylabel('Count')

# Plot 3: Synthetic dataset class distribution
unique_synth, counts_synth = np.unique(synthetic_data['y'], return_counts=True)
axes[1, 0].bar(unique_synth, counts_synth)
axes[1, 0].set_title('Synthetic Dataset - Class Distribution')
axes[1, 0].set_xlabel('Class')
axes[1, 0].set_ylabel('Count')

# Plot 4: Dataset sizes comparison
datasets = ['Digits', 'Iris', 'Wine', 'Synthetic']
samples = [
    digits_data['X'].shape[0],
    iris_data['X'].shape[0],
    sklearn_data['wine']['X'].shape[0],
    synthetic_data['X'].shape[0]
]
axes[1, 1].bar(datasets, samples, color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
axes[1, 1].set_title('Dataset Sizes Comparison')
axes[1, 1].set_ylabel('Number of Samples')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```

## 4. Inject Synthetic Noise

```python
from phase0_preparation.noise_injection import NoiseInjector

# Initialize noise injector
injector = NoiseInjector(random_state=42)

# Use digits dataset for noise injection demonstration
X_digits = digits_data['X']
y_digits = digits_data['y']

# Create noisy versions
noisy_datasets = injector.create_noisy_dataset(
    X_digits, y_digits,
    noise_types=['random', 'class_dependent', 'hard'],
    noise_rates=[0.15, 0.2, 0.1],
    dataset_name='digits'
)

print("Noisy datasets created:")
for name, data in noisy_datasets.items():
    print(f"  • {name}: {data['description']}")
    print(f"    Noise rate: {data['noise_rate']:.3f} ({data['noise_mask'].sum()} noisy samples)")
```

## 5. Visualize Noise Injection Results

```python
# Visualize noise patterns
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Plot clean vs noisy for each noise type
noise_types = ['clean', 'random', 'class_dependent', 'hard']
titles = ['Clean Labels', 'Random Noise (15%)', 'Class-Dependent Noise', 'Hard Example Noise (10%)']

for idx, (noise_type, title) in enumerate(zip(noise_types, titles)):
    ax = axes[idx // 2, idx % 2]
    
    data = noisy_datasets[noise_type]
    y_clean = data.get('y_clean', data['y'])  # Some datasets store clean labels separately
    y_noisy = data['y']
    
    # Plot confusion between clean and noisy labels
    confusion = np.zeros((10, 10))
    for clean_label, noisy_label in zip(y_clean, y_noisy):
        confusion[clean_label, noisy_label] += 1
    
    # Normalize by row
    confusion_norm = confusion / confusion.sum(axis=1, keepdims=True)
    
    im = ax.imshow(confusion_norm, cmap='YlOrRd', vmin=0, vmax=1)
    ax.set_title(title)
    ax.set_xlabel('Noisy Label')
    ax.set_ylabel('Clean Label')
    ax.set_xticks(range(10))
    ax.set_yticks(range(10))
    
    # Add text for diagonal (correct labels)
    for i in range(10):
        for j in range(10):
            if i == j:
                text = ax.text(j, i, f'{confusion_norm[i, j]:.2f}',
                              ha="center", va="center", color="black", fontweight='bold')

plt.colorbar(im, ax=axes, location='right', pad=0.01, label='Fraction of labels')
plt.suptitle('Label Noise Patterns in Digits Dataset', fontsize=16, y=1.02)
plt.tight_layout()
plt.show()
```

## 6. Save Datasets for Future Phases

```python
# Save clean datasets
all_data = {**sklearn_data, 'synthetic': synthetic_data}
prep.save_dataset_to_csv(all_data, base_path='../data/synthetic')

# Save noisy datasets
injector.save_noisy_datasets(noisy_datasets, base_path='../data/synthetic/digits')

print("✓ All datasets saved successfully!")
print("\nFiles created:")
print("  • Clean datasets: data/synthetic/*_dataset.csv")
print("  • Noisy datasets: data/synthetic/digits/*.csv")
print("\nReady for Phase 1: Data Loading & Exploration!")
```
