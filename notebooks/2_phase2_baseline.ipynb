# Phase 2: Noise Detection - Baseline

**Goal**: Detect suspicious labels using simple disagreement analysis

## What we'll do:
1. K-fold cross-validation
2. Compare predictions with labels
3. Compute disagreement score per sample
4. Rank suspicious samples
5. Analyze class-level disagreement patterns

```python
# Import necessary libraries
import sys
import os
sys.path.append('../src')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Set style
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['figure.dpi'] = 100

print("✓ Libraries imported")
```

## 1. Load Phase 2 Data

```python
def load_phase2_data(data_dir='../data/phase2_ready'):
    """Load datasets prepared in Phase 1."""
    import glob
    
    if not os.path.exists(data_dir):
        print(f"Directory not found: {data_dir}")
        return {}
    
    datasets = {}
    csv_files = glob.glob(os.path.join(data_dir, "*_cleaned.csv"))
    
    print(f"Found {len(csv_files)} dataset files in {data_dir}")
    
    for csv_file in csv_files:
        try:
            df = pd.read_csv(csv_file)
            dataset_name = os.path.splitext(os.path.basename(csv_file))[0].replace('_cleaned', '')
            
            feature_columns = [col for col in df.columns if col != 'target']
            X = df[feature_columns].values
            y = df['target'].values
            
            datasets[dataset_name] = {
                'X': X,
                'y': y,
                'feature_names': feature_columns,
                'dataframe': df,
                'n_samples': len(df),
                'n_features': len(feature_columns),
                'n_classes': len(np.unique(y)),
                'filepath': csv_file
            }
            
            print(f"  ✓ {dataset_name}: {len(df)} samples, {len(feature_columns)} features")
            
        except Exception as e:
            print(f"  ✗ Error loading {csv_file}: {e}")
    
    return datasets

# Load data
datasets = load_phase2_data()

if not datasets:
    print("\nCreating sample dataset...")
    from sklearn.datasets import load_digits
    digits = load_digits()
    
    datasets = {
        'digits': {
            'X': digits.data,
            'y': digits.target,
            'feature_names': [f'pixel_{i}' for i in range(digits.data.shape[1])],
            'n_samples': digits.data.shape[0],
            'n_features': digits.data.shape[1],
            'n_classes': len(np.unique(digits.target))
        }
    }
    print(f"  Created digits dataset: {digits.data.shape[0]} samples")

print(f"\nTotal datasets loaded: {len(datasets)}")
```

## 2. Baseline Noise Detection

```python
from phase2_noise_detection.baseline_detection import BaselineNoiseDetector

# Select a dataset for demonstration
dataset_name = list(datasets.keys())[0]
dataset_info = datasets[dataset_name]
X = dataset_info['X']
y = dataset_info['y']

print(f"Dataset: {dataset_name}")
print(f"  • Samples: {len(y):,}")
print(f"  • Features: {X.shape[1]}")
print(f"  • Classes: {len(np.unique(y))}")

# Initialize detector
detector = BaselineNoiseDetector(
    model_type='logistic',
    n_folds=5,
    random_state=42
)

# Compute disagreement scores
print("\nComputing disagreement scores...")
results = detector.compute_disagreement_scores(X, y)

# Display key results
print(f"\nResults:")
print(f"  • Suspicious labels: {results['n_suspicious']} ({results['disagreement_rate']:.1%})")
print(f"  • Average confidence: {results['avg_confidence']:.3f}")
print(f"  • Confidence (suspicious): {results['avg_confidence_suspicious']:.3f}")
print(f"  • Confidence (correct): {results['avg_confidence_correct']:.3f}")
```

## 3. Rank Suspicious Samples

```python
# Rank suspicious samples
print("Ranking suspicious samples...")
suspicious_ranking = detector.rank_suspicious_samples(results, top_k=20)

if not suspicious_ranking.empty:
    print(f"\nTop 20 most suspicious samples (lowest confidence):")
    print(suspicious_ranking[['sample_index', 'predicted_label', 'confidence', 'fold']].head(20).to_string(index=False))
    
    # Visualization: Confidence distribution
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Confidence histogram
    axes[0].hist(results['confidence_scores'], bins=30, alpha=0.7, edgecolor='black')
    axes[0].set_xlabel('Prediction Confidence')
    axes[0].set_ylabel('Frequency')
    axes[0].set_title('Confidence Distribution')
    axes[0].grid(True, alpha=0.3)
    
    # Confidence vs Disagreement
    disagree_mask = results['suspicious_mask']
    axes[1].scatter(results['confidence_scores'][~disagree_mask], 
                   np.zeros((~disagree_mask).sum()), 
                   alpha=0.5, label='Agree', color='green')
    axes[1].scatter(results['confidence_scores'][disagree_mask], 
                   np.ones(disagree_mask.sum()), 
                   alpha=0.5, label='Disagree', color='red')
    axes[1].set_xlabel('Prediction Confidence')
    axes[1].set_ylabel('Disagreement (0=Agree, 1=Disagree)')
    axes[1].set_title('Confidence vs Disagreement')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.show()
else:
    print("No suspicious samples found!")
```

## 4. Class Analysis

```python
# Analyze by class
print("Analyzing disagreement by class...")
class_analysis = detector.analyze_by_class(X, y, results)

if not class_analysis.empty:
    print("\nClass-level disagreement analysis:")
    print(class_analysis[['class', 'n_samples', 'n_suspicious', 'disagreement_rate', 'avg_confidence']].to_string(index=False))
    
    # Visualization: Class-wise disagreement
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    
    # Disagreement rate by class
    bars1 = axes[0].bar(class_analysis['class'], class_analysis['disagreement_rate'], 
                       color='coral', edgecolor='black')
    axes[0].set_xlabel('Class')
    axes[0].set_ylabel('Disagreement Rate')
    axes[0].set_title('Disagreement Rate by Class')
    axes[0].grid(True, alpha=0.3)
    
    for bar, rate in zip(bars1, class_analysis['disagreement_rate']):
        height = bar.get_height()
        axes[0].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{rate:.3f}', ha='center', va='bottom', fontsize=9)
    
    # Confidence by class
    bars2 = axes[1].bar(class_analysis['class'], class_analysis['avg_confidence'], 
                       color='skyblue', edgecolor='black')
    axes[1].set_xlabel('Class')
    axes[1].set_ylabel('Average Confidence')
    axes[1].set_title('Average Confidence by Class')
    axes[1].grid(True, alpha=0.3)
    
    for bar, conf in zip(bars2, class_analysis['avg_confidence']):
        height = bar.get_height()
        axes[1].text(bar.get_x() + bar.get_width()/2., height + 0.01,
                    f'{conf:.3f}', ha='center', va='bottom', fontsize=9)
    
    plt.tight_layout()
    plt.show()
else:
    print("Class analysis not available")
```

## 5. Cross-Validation Analysis

```python
from phase2_noise_detection.cross_validation import CrossValidationAnalyzer

print("Running comprehensive cross-validation analysis...")

cv_analyzer = CrossValidationAnalyzer(n_folds=5, random_state=42)
cv_results = cv_analyzer.run_cv_analysis(X, y, model_type='logistic')

# Visualize CV results
cv_analyzer.visualize_cv_results(cv_results, dataset_name=dataset_name)
```

## 6. Disagreement Analysis

```python
from phase2_noise_detection.disagreement_analysis import DisagreementAnalyzer

print("Running detailed disagreement analysis...")

disagreement_analyzer = DisagreementAnalyzer(random_state=42)
disagreement_report = disagreement_analyzer.generate_disagreement_report(
    X, y, cv_results, dataset_name=dataset_name
)

print("\nSummary:")
for key, value in disagreement_report['summary'].items():
    print(f"  {key}: {value}")

print("\nRecommendations:")
for i, rec in enumerate(disagreement_report['recommendations'], 1):
    print(f"  {i}. {rec}")

# Visualize disagreement analysis
disagreement_analyzer.visualize_disagreement_analysis(X, y, cv_results, dataset_name=dataset_name)
```

## 7. Save Results

```python
# Create output directory
output_dir = '../reports/phase2_notebook'
os.makedirs(output_dir, exist_ok=True)

# Save suspicious samples
if not suspicious_ranking.empty:
    suspicious_file = os.path.join(output_dir, f'{dataset_name}_suspicious_samples.csv')
    suspicious_ranking.to_csv(suspicious_file, index=False)
    print(f"✓ Suspicious samples saved to {suspicious_file}")

# Save class analysis
if not class_analysis.empty:
    class_file = os.path.join(output_dir, f'{dataset_name}_class_analysis.csv')
    class_analysis.to_csv(class_file, index=False)
    print(f"✓ Class analysis saved to {class_file}")

# Create summary
summary = {
    'dataset': dataset_name,
    'n_samples': len(y),
    'n_features': X.shape[1],
    'n_classes': len(np.unique(y)),
    'n_suspicious': results['n_suspicious'],
    'disagreement_rate': results['disagreement_rate'],
    'overall_accuracy': cv_results['overall_accuracy'],
    'overall_confidence': cv_results['overall_confidence'],
    'timestamp': pd.Timestamp.now().isoformat()
}

import json
summary_file = os.path.join(output_dir, f'{dataset_name}_summary.json')
with open(summary_file, 'w') as f:
    json.dump(summary, f, indent=2)

print(f"✓ Summary saved to {summary_file}")

print("\n" + "="*60)
print("PHASE 2 COMPLETE!")
print("="*60)
print(f"\nAnalyzed dataset: {dataset_name}")
print(f"• Samples: {len(y):,}")
print(f"• Suspicious labels detected: {results['n_suspicious']} ({results['disagreement_rate']:.1%})")
print(f"• Overall accuracy: {cv_results['overall_accuracy']:.3f}")
print(f"• Overall confidence: {cv_results['overall_confidence']:.3f}")
print(f"\nOutput saved to: {output_dir}/")
```
